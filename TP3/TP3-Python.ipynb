1) Python permite la generacion de numeros aleatorios de manera nativa utilizando ```random``` o haciendo uso de la libreria ```numpy```.

````python
import random # Importar modulo

num_random = [random.randint(0, 100) for _ in range(100)]
print(num_random)
````

El modulo ```random``` implementa generadores de números pseudoaleatorios para varias distribuciones.
Para los enteros, existe una selección uniforme dentro de un rango. Para las secuencias, existe una selección uniforme de un elemento aleatorio, una función para generar una permutación aleatoria de una lista in-situ y una función para el muestreo aleatorio sin reemplazo.
Para números reales, existen funciones para calcular distribuciones uniformes, normales (Gaussianas), log-normales, exponenciales negativas, gamma y beta. Para generar distribuciones angulares está disponible la distribución de von Mises.

````python
import numpy as np # Importar la libreria

num_numpy = np.random.randint(0, 100, 100)
print(num_numpy)
````

La libreria ```numpy``` proporciona un objeto de matriz multidimensional, varios objetos derivados (como matrices y matrices enmascaradas) y una variedad de rutinas para operaciones rápidas en matrices, incluidas matemáticas, lógicas, manipulación de formas, clasificación, selección, etc. La funcion que utilizaremos es numpy.random para generar un arreglo psudoaleatorio de numeros. 

Tambien podemos generar numeros aleatorios siguiendo una distribucion probabilistica normal de la siguiente manera utilizando random.normal(loc=0, scale=1, size=X) Donde:

Los parametros de entrada son:
* loc: media de la distribucion
* scale: el desvio estandar
* size: la cantidad de muestras

Y retorna muestras extraídas de la distribución normal parametrizada.

````python
muestra = np.random.normal(loc=0, scale=0.1, size=100)
print(muestra)
````

Podremos calcular la media, el desvio estandar y la varianza utilizando las funciones ```numpy.mean()``` para calcular la media,```numpy.std()``` para calcular el desvio estandar y ```numpy.var()``` para calcular la varianza de la siguiente manera:


a) Calcula la media.

````python
media = np.mean(muestra)
print("Valor de la media: ",media)
````

b) Calcular el desvio estandar

````python
desvio = np.std(muestra)
print("Valor del desvio estandar: ", desvio)
````

c) Calcular la varianza

````python
varianza = np.var(muestra)
print("Valor de la varianza: ", varianza)
````
